```python
m = 0 # slop
b = 0 # bias
def personal_linear_regression(x_train,y_train):
    numerator = 0
    denominator = 0
    
    print(x_train.shape,y_train.shape)
    
    for i in range(x_train.shape[0]):
        numerator += (x_train[i] - np.mean(x_train)) * (y_train[i] - np.mean(y_train))
        denominator += (x_train[i] - np.mean(x_train))**2
    
    m = numerator/denominator
    b = np.mean(y_train) - m*np.mean(x_train)
    print(m , b)
    
    return m,b
```

```python
m,b = personal_linear_regression(x_train,y_train)
```
